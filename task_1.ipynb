```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = "/content/drive/MyDrive/world_population task_1.csv"
df = pd.read_csv(file_path)

# Aggregate population by continent
continent_population = df.groupby("Continent")["2022 Population"].sum().reset_index()

# Plot the bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x="Continent", y="2022 Population", data=continent_population, palette="viridis")
plt.xlabel("Continent")
plt.ylabel("Total Population in 2022")
plt.title("Population Distribution by Continent (2022)")
plt.xticks(rotation=45)
plt.show()
```
