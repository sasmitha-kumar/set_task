```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = "/content/drive/MyDrive/accident dataset task_4.csv"  # Ensure correct file path
df = pd.read_csv(file_path)

# Display basic info
print(df.info())

# Check for missing values
print(df.isnull().sum())

# Analyzing Accident Patterns Based on Light Conditions
plt.figure(figsize=(10, 5))
sns.countplot(data=df, x="Light_conditions", order=df["Light_conditions"].value_counts().index, palette="viridis")
plt.xticks(rotation=45)
plt.xlabel("Light Conditions")
plt.ylabel("Number of Accidents")
plt.title("Accident Distribution Based on Light Conditions")
plt.show()


# Display basic info
print(df.info())
print(df.isnull().sum())

# Set style
sns.set_style("whitegrid")

### 1. Accidents by Road Surface Type ###
plt.figure(figsize=(10, 5))
sns.countplot(data=df, x="Road_surface_type", order=df["Road_surface_type"].value_counts().index, palette="coolwarm")
plt.xticks(rotation=45)
plt.xlabel("Road Surface Type")
plt.ylabel("Number of Accidents")
plt.title("Accidents Based on Road Surface Type")
plt.show()

### 2. Accidents by Weather Conditions ###
plt.figure(figsize=(10, 5))
sns.countplot(data=df, x="Weather_conditions", order=df["Weather_conditions"].value_counts().index, palette="magma")
plt.xticks(rotation=45)
plt.xlabel("Weather Conditions")
plt.ylabel("Number of Accidents")
plt.title("Accidents Based on Weather Conditions")
plt.show()

### 3. Accidents by Light Conditions (Time of Day) ###
plt.figure(figsize=(10, 5))
sns.countplot(data=df, x="Light_conditions", order=df["Light_conditions"].value_counts().index, palette="viridis")
plt.xticks(rotation=45)
plt.xlabel("Light Conditions")
plt.ylabel("Number of Accidents")
plt.title("Accidents Based on Light Conditions (Time of Day)")
plt.show()

### 4. Heatmap of Correlations (if numerical data exists) ###
plt.figure(figsize=(8, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()






```
