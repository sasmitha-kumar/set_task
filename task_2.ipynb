```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
file_path = "/content/drive/MyDrive/Titanic-Dataset  task_2.csv"
df = pd.read_csv(file_path)

# Display basic information
display(df.head())  # First few rows
display(df.info())  # Column info
display(df.describe())  # Summary statistics

# Check for missing values
missing_values = df.isnull().sum()
print("Missing Values:\n", missing_values)

# Handle missing values (example: filling missing age with median)
df['Age'].fillna(df['Age'].median(), inplace=True)
df.drop(columns=['Cabin'], inplace=True)  # Dropping 'Cabin' if too many missing values

# Visualizing survival counts
sns.countplot(x='Survived', data=df)
plt.title('Survival Count')
plt.show()

# Analyzing survival rate by gender
sns.barplot(x='Sex', y='Survived', data=df)
plt.title('Survival Rate by Gender')
plt.show()

# Age distribution
sns.histplot(df['Age'], bins=30, kde=True)
plt.title('Age Distribution')
plt.show()

# Checking correlation
plt.figure(figsize=(10,6))
sns.heatmap(df.select_dtypes(include=['number']).corr(), annot=True, cmap='coolwarm')
plt.title('Feature Correlation Heatmap')
plt.show()
```
